```python
midwest = pd.read_csv('midwest.csv')
midwest
```

```python
midwest = midwest.assign(perkid = (1 - (midwest['popadults'] / midwest['poptotal']))*100)
```

```python
midwest
```

```python
midwest.sort_values('perkid', ascending = False)[['county', 'perkid']].head(5)
```

```python
midwest = midwest.assign(grade_perkid = np.where(midwest['perkid'] >= 40, 'large', \
                                                np.where(midwest['perkid'] >= 30, 'middle', 'small')))
midwest
```

```python
midwest.assign(perasian = midwest['popasian'] / midwest['poptotal'] * 100)\
.sort_values('perasian')[['state', 'county', 'perasian']]
```

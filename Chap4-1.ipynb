```python
df = pd.DataFrame({'제품':['사과', '딸기', '수박'],
                  '가격':[1800,1500,3000],
                  '판매량':[24,38,13]})
```

```python
fruit_price_avg = sum(df['가격'])//3
fruit_sale_avg = sum(df['판매량'])//3
print('과일 평균 가격  :' , fruit_price_avg)
print('과일 평균 판매량 :' , fruit_sale_avg)
```

```python
import numpy as np
```

```python
df_mpg['test'] = np.where(df_mpg['total'] >= 20, 'Pass', 'Fail')
df_mpg.head()
```

```python
df_mpg['test'].value_counts()
```

```python
df_mpg_count = df_mpg['test'].value_counts()
```

```python
df_mpg_count.plot.bar()
```

```python
df_mpg_count.plot.bar(rot=0)
```

```python
df_mpg['grade'] = np.where(df_mpg['total'] >= 30, 'A',
                          np.where(df_mpg['total'] >= 20, 'B', 'C'))

df_mpg.head()
```

```python
count_grade = df_mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```

```python
count_grade = df_mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```
